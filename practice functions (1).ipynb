# Formal and Actual Arguments

```python
def person(name,age):    # Formal Argument
    print(name)
    print(age)
person('nit',23)          #Actual Argument
```

```python
def person(name,age):
    print(name)
    print(age)
person(23,'nit')
```

# Keyword argument

```python
def person(name,age):
    print(name)
    print(age+1)
person(age=23,name='nit')
```

```python
def person (name,age1,city):
    print(name)
    print(age1+1)
    print(city)
person(age1=23,name='nit',city='hyd')
```

# wap to count the even or odd number

```python
def count(lst):
    even=0
    odd=0
    for i in lst:
        if i % 2==0:
            even +=1
        else:
            odd +=1
    return even,odd
lst=[1,2,3,4,5,6,7,8,9]
even,odd=count(lst)
print(even)
print(odd)
```

# Fibonacci series

```python
def fib(n):
    a=0
    b=1
    print(a)
    print(b)
    for i in range(0,n):#These process is continue so we use the loop
        c=a+b  
        a=b
        b=c
        print(c)
fib(10)
```
