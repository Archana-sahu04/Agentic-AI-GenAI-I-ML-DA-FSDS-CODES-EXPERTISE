# factorial using recursiom

```python
def fact(n):
    if n==0:
        return 1
    return n* fact(n-1)
result=fact(5)
result
```

```python
def is_odds(n):
    return n%2!=0
    nums=[3,2,6,8,4,6,2,9]
    odds=list(filter(is_odds,nums))
print(odds)
```
